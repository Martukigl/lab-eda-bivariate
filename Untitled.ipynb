```python
import pandas as pd
import numpy as np
import scipy.stats as st
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('amz_uk_price_prediction_dataset-Copy1.csv')
```

```python
print(df.columns)
```

```python
# 1. Crosstab Analysis
# Create crosstab to see counts
ct = pd.crosstab(df['category'], df['isBestSeller'])

# Calculate proportion of best-sellers per category
# We normalize by 'index' (rows) to get the % of BestSellers within each category
prop = pd.crosstab(df['category'], df['isBestSeller'], normalize='index')

# Sorting to find where being a best-seller is more prevalent
best_seller_trends = prop[True].sort_values(ascending=False)

print("Top 10 Categories by Best-Seller Proportion:")
print(best_seller_trends.head(10))

# 2. Statistical Tests
# Chi-square test
chi2, p_val, dof, expected = st.chi2_contingency(ct)

# Cram√©r's V Calculation (Strength of association)
n = ct.sum().sum()
min_dim = min(ct.shape) - 1
cramers_v = np.sqrt(chi2 / (n * min_dim))

print(f"\nChi-square P-value: {p_val:.4f}")
print(f"Cram√©r's V: {cramers_v:.4f}")

# 3. Visualizations
# Filtering the top 20 categories by best-seller proportion for a cleaner chart
top_20_categories = best_seller_trends.head(20).index
prop_filtered = prop.loc[top_20_categories]

prop_filtered.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#ff9999','#66b3ff'])
plt.title('Best-Seller Status Proportion by Category (Top 20)')
plt.ylabel('Proportion')
plt.xlabel('Category')
plt.legend(title='Is Best Seller?', labels=['No', 'Yes'])
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Calculate IQR
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the dataframe
df_clean = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)].copy()

print(f"Original dataset size: {len(df)}")
print(f"Cleaned dataset size: {len(df_clean)}")
print(f"Outliers removed: {len(df) - len(df_clean)}")
```

```python
# 1. Violin Plot (Top 20 categories by count)
top_20_cats = df_clean['category'].value_counts().nlargest(20).index
df_top20 = df_clean[df_clean['category'].isin(top_20_cats)]

plt.figure(figsize=(12, 8))
sns.violinplot(data=df_top20, x='category', y='price')
plt.xticks(rotation=90)
plt.title('Price Distribution - Top 20 Categories')
plt.show()

# Answer: Category with highest median price (without filtering)
median_prices = df_clean.groupby('category')['price'].median().sort_values(ascending=False)
print(f"Category with highest median price: {median_prices.index[0]}")

# 2. Bar Chart (Average Price - Top 10 categories by count)
top_10_cats = df_clean['category'].value_counts().nlargest(10).index
df_top10 = df_clean[df_clean['category'].isin(top_10_cats)]

plt.figure(figsize=(10, 6))
sns.barplot(data=df_top10, x='category', y='price', estimator=np.mean)
plt.xticks(rotation=45)
plt.title('Average Price - Top 10 Categories')
plt.show()

# Answer: Category with highest average price
avg_prices = df_clean.groupby('category')['price'].mean().sort_values(ascending=False)
print(f"Category with highest average price: {avg_prices.index[0]}")

# 3. Box Plots (Ratings - Top 10 categories by count)
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top10, x='category', y='stars')
plt.xticks(rotation=45)
plt.title('Ratings Distribution - Top 10 Categories')
plt.show()

# Answer: Category with highest median rating
median_ratings = df_clean.groupby('category')['stars'].median().sort_values(ascending=False)
print(f"Category with highest median rating: {median_ratings.index[0]}")
```

```python
# 1. Correlation Coefficient
correlation = df_clean['price'].corr(df_clean['stars'])
print(f"Correlation between Price and Stars: {correlation:.4f}")

# 2. Visualizations
# Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_clean, x='stars', y='price', alpha=0.3)
plt.title('Price vs. Ratings (Stars)')
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df_clean.select_dtypes(include=[np.number]).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# QQ Plot (Normal distribution check for Price)
import statsmodels.api as sm
plt.figure(figsize=(8, 6))
sm.qqplot(df_clean['price'], line='s')
plt.title('QQ Plot for Price Distribution')
plt.show()
```
